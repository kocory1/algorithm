```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def basic_quicksort(arr):
    if len(arr) <= 1 :
        return arr
    pivot = arr[len(arr)//2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if  x == pivot]
    right = [x for x in arr if x> pivot]
    return basic_quicksort(left) + middle + basic_quicksort(right)

arr = [5, 2, 8, 6, 0, 7, 1, 4, 9, 3]
print(basic_quicksort(arr))
```

```python
def partition(arr, p, r):
    x = arr[r]
    i = p - 1
    for j in range(p, r):
        if arr[j] <= x:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i+1], arr[r] = arr[r], arr[i+1]
    return i+1

arr = [5, 2, 8, 6, 0, 7, 1, 4, 9, 3]
p = 0
r = len(arr) - 1
print(partition(arr, p, r))
```

```python
def quicksort(arr,p,r):
    if p < r:
        q = partition(arr, p, r)
        quicksort(arr, p, q-1)
        quicksort(arr, q+1, r)

arr = [5, 2, 8, 6, 0, 7, 1, 4, 9, 3]
quicksort(arr, 0, len(arr)-1)

print(arr)
```

```python
def randomized_partition(arr, p, r):
    i = np.random.randint(p, r+1)
    arr[r], arr[i] = arr[i], arr[r]
    return partition(arr, p, r)

def randomized_quicksort(arr, p, r):
    if p < r:
        q = randomized_partition(arr, p, r)
        randomized_quicksort(arr, p, q-1)
        randomized_quicksort(arr, q+1, r)

nums = [1000, 10000, 100000, 1000000]

quicksort_time = []
randomized_quicksort_time = []

for n in nums:
    x = np.random.permutation(n)
    start = time.time()
    quicksort(x, 0, len(x)-1)
    end = time.time()
    quicksort_time.append(end-start)
    
    x = np.random.permutation(n)
    start = time.time()
    randomized_quicksort(x, 0, len(x)-1)
    end = time.time()
    randomized_quicksort_time.append(end-start)
    
plt.plot(nums, quicksort_time, 'o-', label='quicksort')
plt.plot(nums, randomized_quicksort_time, 'o-', label='randomized quicksort')
plt.legend()
plt.xlabel('number of elements')
plt.ylabel('time')
plt.title('quicksort and randomized quicksort')
plt.show()
```
