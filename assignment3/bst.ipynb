```python
class TreeNode:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None
        self.parent = None

class BST:
    def __init__(self):
        self.root = None
    
    def insert(self, node):
        x = self.root
        y = None
        while x != None:
            y = x
            if node.key < x.key:
                x = x.left
            else:
                x = x.right
        node.parent = y
        if y is None:
            self.root = node
        elif node.key < y.key:
            y.left = node
        else:
            y.right = node
            
    def find(self, key):
        x = self.root
        if key == "root":
            return x
        while x is not None and x.key != key:
            if x.key > key:
                x = x.left
            else:
                x = x.right
        return x
    
    def minimum(self, node):
        x = self.find(node)
        while x.left is not None:
            x = x.left
        return x
    
    def transplant(self, u, v):
        if u.parent is None:
            self.root = v
            return 
        elif u == u.parent.left:
            u.parent.left = v
        else :
            u.parent.right = v
        
        if v is not None:
            v.parent = u.parent
        
    def delete(self, key):
        x = self.find(key)
        if x.left is None:
            self.transplant(x, x.right)
        elif x.right is None:
            self.transplant(x, x.left)
        else:
            y = self.minimum(x.right)
            if y != x.right:
                self.transplant(y, y.right)
                y.right = x.right
                y.right.parent = y
            self.transplant(x, y)
            y.left = x.left
            y.left.parent = y
            
    def inorder(self, node):
        if node is not None:
            self.inorder(node.left)
            print(node.key, ' ')
            self.inorder(node.right)
    
    def display(self):
        self.inorder(self.root)
        
```

```python
numbers = {23, 15, 10, 41, 20, 16, 35, 28, 17, 14, 39, 30, 3, 7, 12, 38, 26, 19, 47, 21}

bst = BST()
for num in numbers:
    node = TreeNode(num)
    bst.insert(node)    
bst.display()
```

```python
d_node = {35, 17, 41, 'root'}
for key in d_node:
    bst.delete(key)
bst.display()
```
